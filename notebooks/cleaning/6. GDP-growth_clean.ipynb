# Gross Domestic Product (GDP) growth rate - Data Cleaning 

### Data Info:
* **Indicator:** `Real GDP growth rate (yoy)`
* **Source:** National Anti-Corruption Bureau of Ukraine: https://nabu.ua/ua/vvp-2.html 
* **Raw File:** `data_raw/GDP.xlsx`
* **Output:** `data_processed/GDP_clean.xlsx`  

### Cleaning Notes:
- source file contains multi-row headers and merged cells

```python
import pandas as pd
import numpy as np
import re
from pathlib import Path
```

```python
file_path = Path("../../data_raw/GDP.xls")
output_path = Path("../../data_processed/GDP_clean.xlsx")
output_path.parent.mkdir(parents=True, exist_ok=True)

sheet_name = "Sheet1 "

df_raw = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
df_raw.head()
```

```python
# set correct header, only keep needed columns

df = df_raw.copy()
df = df.iloc[:, [1, 3]].copy()

df.columns = ["date", "value"]
df = df.iloc[5:].copy()

df.head(5)
```

```python
# clean quarter labels
# 1 кв. 2000 -> 2000Q1

df["date"] = (
    df["date"]
    .astype(str)
    .str.strip()
    .str.replace(r"(\d)\s*кв\.\s*(\d{4})", r"\2Q\1", regex=True)
)


df["indicator"] = "gdp_growth"

df["year"] = df["date"].str[:4].astype(int)
df = df[df["year"] >= 2015].copy()
df = df.drop(columns="year")

df.head(10)
```

```python
df.to_excel(output_path, index=False)
print(f"saved to: {output_path}")
```
